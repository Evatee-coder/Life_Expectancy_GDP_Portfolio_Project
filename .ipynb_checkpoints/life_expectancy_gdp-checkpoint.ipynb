### Explore the distribution of GDP and Life expectancy at birth (years)

* Below the distribution of GDP is shown.

```python
plt.figure(figsize=(8,6))
sns.histplot(df.GDP)
plt.xlabel("GDP in Trillions of U.S. Dollars");
plt.title('Distribution of GDP')
```

As seen from the above distribution of GDP, the distribution is right skewed as most of the value concentrate on the left. The right-skewed distribution is also known as a positively skewed distribution, which indicates that the tail of the distribution is skewed towards the higher GDP values. In the context of GDP  where the distribution is right-skewed and concentrates below 0.25, it suggests the following:

*  Dominance of Lower GDPs: The concentration of the distribution below 0.25 suggests that a majority of the countries have GDP values that are relatively lower. This could indicate that there are many smaller economies or entities with limited economic output.

* Developing or Less Developed Economies: A right-skewed distribution in the context of GDP could point towards a scenario where many economies are still developing or are characterized as less developed. These economies might struggle with various factors like limited access to resources, lower industrialization, and lower levels of economic development.

* Disparities in Economic Growth: The distributions skewness suggests that there might be a few entities or regions with significantly higher GDP values, contributing to the skewness. This could be due to factors such as resource abundance, advanced technology, favorable policies, or other drivers of rapid economic growth.
 

* Exploring the distribution of Life expectancy at birth (years)

```python
plt.figure(figsize = (8,6))
sns.histplot(df.LEABY)
plt.xlabel("Life expectancy at birth (years)");
plt.title('Distribution of Life expectancy at birth (years)')
```

In contrast to the GDP distribution, the distribution of life expectancy at birth (years) is a left-skewed distribution, also known as a negatively skewed distribution as most of its values are concentrated on the right side of the distribution. This indicates that the tail of the distribution is skewed towards the lower values. In the context of life expectancy at birth, as the values concentrates above 70 years suggest the following:

* Higher Life Expectancy: The fact that the distribution is concentrated above 70 years indicates that a significant portion of the population has a life expectancy greater than 70 years. This could be indicative of improvements in healthcare, nutrition, living conditions, and overall quality of life that contribute to increased life expectancy.

* Aging Population: The concentration of the distribution above 70 years suggests an aging population. This could be due to factors such as lower birth rates, increased life expectancy of the elderly, and improvements in healthcare for the elderly population. An aging population has implications for healthcare systems, social security programs, and other services geared towards older individuals.

* Healthcare and Lifestyle Improvements: The left-skewed distribution could reflect improvements in healthcare access, medical advancements, disease prevention, and healthier lifestyles. These factors contribute to a greater number of people living longer lives.

#### To further confirm if there is a correlation between GDP and life expectancy of a country?

* We made further plot below fitted with a liner curve.

```python
# Is there a correlation between GDP and life expectancy of a country?

countries =  df.Country.unique()
colors = ['red', 'purple', 'orange', 'darkblue', 'violet', 'green']
# Below we create a figure with 2 rows and 3 columns
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(16, 12))         
fig.suptitle('Correlation between Life Expectancy and GDP', size=20) 
# Below we create loop for scatterplots for each country
for i, country in enumerate(countries):
    # Below remainder of division (column position)
    cols = i % 3 
    # integer division (row position)
    rows = i // 3     
    # below figure position
    ax = axes[rows, cols]                                           
    country_df = df[df.Country == country]    
    # Below, we create a start for x labels
    x_values = [x for x in range(                                   
        len(country_df['LEABY']))]
     # Given that x-axis value are floats, below we round the values
    rounded_values = [                                             
        round(value) for value in country_df['LEABY']]
    # Using regplot to show the correlation between Life Expectancy and GDP. 
    sns.regplot(data=country_df, ax=ax,                           
        x='LEABY', y='GDP', 
        color=colors[i])    # Each Country has its own color 
    # Below we replace the start x labels with rounded values
    ax.set_xticks(rounded_values)   
    # The code below is to create title and labels
    ax.set_title(f'{country}')                                      
    ax.set_xlabel('Life expectancy at birth (years)')
    ax.set_ylabel('GDP')
    
    # Calculate the correlation coefficient
    corr_coeff, _ = pearsonr(country_df['LEABY'], country_df['GDP'])
    
    # Add the correlation coefficient value to the plot
    ax.text(0.15, 0.9, f'Corr: {corr_coeff:.2f}', transform=ax.transAxes, fontsize=12,
            ha='center', va='center', bbox=dict(facecolor='white', edgecolor='black', boxstyle='round'))
     
     
# Adjust the position and size of the charts on the figure
plt.tight_layout(rect=[0, 0.05, 1, 0.95])
plt.savefig('figures\Correlation between Life Expectancy and GDP.png')
plt.show()
plt.clf()
```

It is evident from the correlations that all the GDP and Life expectancy in all countries are strongly correlated 

### Violin plot

* The Violin plot offers an additional way to visualize the distributions of individual elements and identify patterns in their shapes. This type of plot is widely favored due to its ability to showcase the distribution's shape in comparison to a traditional box plot.

```python
fig, axes = plt.subplots(1, 2, sharey=True, figsize=(15, 5))
axes[0] = sns.violinplot(ax=axes[0], x=df.GDP, y=df.Country)
axes[0].set_xlabel("GDP in Trillions of U.S. Dollars")
axes[1] = sns.violinplot(ax=axes[1], x=df.LEABY, y=df.Country)
axes[1].set_xlabel("Life expectancy at birth (years)");
```

* In the plots shown above, the countries are represented on the vertical y-axis, while the numerical columns for GDP and life expectancy are on the horizontal x-axis. In the left plot (country vs. GDP), the USA and China exhibit a broader range, whereas Zimbabwe, Chile, and Mexico have more constrained ranges.

* In the right plot which displays life expectancy, most countries have narrower ranges, except for Zimbabwe, that displays a broader range that spans from the upper 30s to the upper 60s.

### Swarm Plots

* A more recent approach to illustrating distributions involves using swarm plots, which can be employed alongside box and violin plots to provide additional insights. Initially, we present the standalone swarm plot, followed by its overlay on a violin plot. Swarm plots are valuable due to their ability to display the density of dots around specific values, offering a visual representation of distribution through both dot density and the shape of the area.

```python
# Define a color palette for the countries
country_palette = sns.color_palette("Set1", n_colors=len(df.Country.unique()))

fig, axes = plt.subplots(1, 2, sharey=True, figsize=(15, 5))

# Left plot (GDP)
axes[0] = sns.swarmplot(ax=axes[0], x=df.GDP, y=df.Country, palette=country_palette)  # Use the defined palette
axes[0].set_xlabel("GDP in Trillions of U.S. Dollars")

# Right plot (Life expectancy)
axes[1] = sns.swarmplot(ax=axes[1], x=df.LEABY, y=df.Country, palette=country_palette)  # Use the defined palette
axes[1].set_xlabel("Life expectancy at birth (years)")

plt.tight_layout()
plt.show()
```

```python
country_palette = sns.color_palette("Set1", n_colors=len(df.Country.unique()))

fig, axes = plt.subplots(1, 2, sharey=True, figsize=(15, 5))

# Left plot (GDP)
axes[0] = sns.violinplot(ax=axes[0], x=df.GDP, y=df.Country, color="black")
axes[0] = sns.swarmplot(ax=axes[0], x=df.GDP, y=df.Country, palette=country_palette)  # Use the defined palette
axes[0].set_xlabel("GDP in Trillions of U.S. Dollars")

# Right plot (Life expectancy)
axes[1] = sns.violinplot(ax=axes[1], x=df.LEABY, y=df.Country, color="black")
axes[1] = sns.swarmplot(ax=axes[1], x=df.LEABY, y=df.Country, palette=country_palette)  # Use the defined palette
axes[1].set_xlabel("Life expectancy at birth (years)")

plt.tight_layout()
plt.show()
```

* When examining the GDP plot on the left side, Chile and Zimbabwe are characterized by a vertical line of dots that effectively depict the concentration of data points around their respective values. This level of detail might have been overlooked if using only a box plot, unless the viewer possesses a high level of proficiency in interpreting data visualizations.
